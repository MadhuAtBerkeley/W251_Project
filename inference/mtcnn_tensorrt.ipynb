```python
import torch
import torch.onnx
from models.mtcnn import PNet, RNet, ONet
import tensorrt
```

```python
BATCH_SIZE = 1
H = 700
W = int(1280*12/40)
USE_FP16 = 1
pnet = PNet()
dummy_input=torch.randn(1, 3, H, W)
torch.onnx.export(pnet, dummy_input, "pnet1.onnx", verbose=False)


# step out of Python for a moment to convert the ONNX model to a TRT engine using trtexec
if USE_FP16:
   !/usr/src/tensorrt/bin/trtexec --onnx=pnet.onnx --saveEngine=pnet_engine1.trt  --explicitBatch --inputIOFormats=fp16:chw --outputIOFormats=fp16:chw --fp16
else:
   !/usr/src/tensorrt/bin/trtexec --onnx=pnet.onnx --saveEngine=pnet_engine.trt  --explicitBatch
```
